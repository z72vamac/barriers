```python
import numpy as np

from scipy.spatial import Delaunay

points = np.random.uniform(0, 100, (10, 2))

points

```

```python
tri = Delaunay(points)

def baricenter(points, tri):
    
    P = []
    
    for a, b, c in tri.simplices:
        P = (points[a, :] + points[b, :] + points[c, :])/3
    
    return P


    
```
